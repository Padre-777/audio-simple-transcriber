```python
import os
from google.colab import files

# ==============================================================================
# 1. УСТАНОВКА НЕОБХОДИМЫХ БИБЛИОТЕК (всегда в самом начале!)
# ==============================================================================
print("📦 Устанавливаю Whisper и ffmpeg...")
!pip install -q git+https://github.com/openai/whisper.git
# !apt-get -qq install ffmpeg

# Теперь можно импортировать whisper после установки
import whisper

# ==============================================================================
# 2. РЕЖИМ 1: ЗАГРУЗКА НОВОГО ФАЙЛА С КОМПЬЮТЕРА
# ==============================================================================
print("📁 Выберите файл на вашем компьютере")
uploaded = files.upload()
audio_file = list(uploaded.keys())[0]

# ==============================================================================
# 2. (Альтернатива) РЕЖИМ 2: ИСПОЛЬЗОВАНИЕ УЖЕ ЗАГРУЖЕННОГО ФАЙЛА
# Раскомментируйте следующие 2 строки и закомментируйте 2 строки выше
# ==============================================================================
# audio_file = "выпуск 1_2605.mp3"
# uploaded = None  # чтобы не вызывать files.upload()

# ==============================================================================
# 3. РАСШИФРОВКА
# ==============================================================================
if audio_file and os.path.exists(audio_file):
    print(f"\n🎤 Расшифровываю {audio_file} (Размер: {os.path.getsize(audio_file)/1024/1024:.1f} МБ)...")

    # Загружаем модель (можно выбрать 'tiny', 'base', 'small', 'medium', 'large')
    model = whisper.load_model("medium")
    result = model.transcribe(audio_file, language="ru")

    # Сохраняем результат
    with open("transcription.txt", "w", encoding="utf-8") as f:
        f.write(result["text"])

    print(f"\n✅ Готово! Текст сохранён в transcription.txt")
    print(f"\n📝 Первые 500 символов:\n{result['text'][:500]}")
else:
    print(f"❌ Файл '{audio_file}' не найден. Проверьте имя или загрузите файл.")
```
