```python
import os
import whisper
from google.colab import files

# ==============================================================================
# РЕЖИМ 1: ЗАГРУЗКА НОВОГО ФАЙЛА С КОМПЬЮТЕРА
# (Раскомментируйте этот блок, если нужно загрузить новый файл)
# ==============================================================================
# print("📁 Выберите файл на вашем компьютере")
# uploaded = files.upload()
# # Автоматически берем имя только что загруженного файла
# audio_file = list(uploaded.keys())[0]


# ==============================================================================
# РЕЖИМ 2: ИСПОЛЬЗОВАНИЕ УЖЕ ЗАГРУЖЕННОГО В COLAB ФАЙЛА
# (Раскомментируйте этот блок, если файл уже лежит в папке /content/)
# ==============================================================================
audio_file = "выпуск 1_2605.mp3"


# ==============================================================================
# БЛОК УСТАНОВКИ И РАСШИФРОВКИ (Остается всегда активным)
# ==============================================================================

# Устанавливаем Whisper и ffmpeg (если ещё не установлены)
!pip install -q git+https://github.com
!apt-get -qq install ffmpeg

# Проверяем, существует ли указанный файл, и запускаем процесс
if audio_file and os.path.exists(audio_file):
    print(f"\n🎤 Расшифровываю {audio_file} (Размер: {os.path.getsize(audio_file)/1024/1024:.1f} МБ)...")

    # Загружаем модель (base — компромисс скорости и качества, turbo — точнее)
    model = whisper.load_model("medium")
    result = model.transcribe(audio_file, language="ru")

    # Сохраняем результат
    with open("transcription.txt", "w", encoding="utf-8") as f:
        f.write(result["text"])

    print(f"\n✅ Готово! Текст сохранён в transcription.txt")
    print(f"\n📝 Первые 500 символов:\n{result['text'][:500]}")
else:
    print(f"❌ Файл '{audio_file}' не найден. Проверьте имя или загрузите файл в боковую панель.")
```
