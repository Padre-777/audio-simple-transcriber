```python
# 1. Загружаем файл с компьютера
from google.colab import files
print("📁 Выберите файл audio.aac на вашем компьютере")
uploaded = files.upload()

# 2. Проверяем, что файл загрузился
import os
print("\n📂 Файлы в текущей папке:")
for filename in os.listdir():
    if filename.endswith('.aac') or filename.endswith('.mp3') or filename.endswith('.wav'):
        print(f"   ✅ {filename} - {os.path.getsize(filename)/1024/1024:.1f} МБ")

# 3. Устанавливаем Whisper (если ещё не установлен)
!pip install -q git+https://github.com/openai/whisper.git
!apt-get -qq install ffmpeg

# 4. Расшифровываем (поддерживает .m4a, .aac, .mp3, .wav)
import whisper

# Находим любое подходящее аудио
audio_extensions = ('.aac', '.m4a', '.mp3', '.wav', '.flac')
audio_file = None

for f in os.listdir():
    if f.lower().endswith(audio_extensions):
        audio_file = f
        break

if audio_file:
    print(f"\n🎤 Расшифровываю {audio_file}...")
    model = whisper.load_model("medium")
    result = model.transcribe(audio_file, language="ru")

    # Сохраняем результат
    with open("transcription.txt", "w", encoding="utf-8") as f:
        f.write(result["text"])

    print(f"\n✅ Готово! Текст сохранён в transcription.txt")
    print(f"\n📝 Первые 500 символов:\n{result['text'][:500]}")
else:
    print("❌ Поддерживаемые аудиофайлы не найдены. Убедитесь, что загрузили нужный файл.")
```
